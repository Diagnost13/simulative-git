#### **Задача 1**

Ваша задача написать класс `Extraction`, и определить в нем метод `from_csv`, который принимал бы путь к файлу и вытаскивал данные в виде списка объектов.

Для этого, здесь же, определите класс `Sale`.

**Решение**

Напишите свое решение ниже

```python
import csv

class Sale:
    def __init__(self, id, date, amount, product):
        self._id = id
        self._date = date
        self._amount = amount
        self._product = product

    def get_id(self):
        return self._id

    def set_id(self, value):
        self._id = value

    def get_date(self):
        return self._date

    def set_date(self, value):
        self._date = value

    def get_amount(self):
        return self._amount

    def set_amount(self, value):
        self._amount = value

    def get_product(self):
        return self._product

    def set_product(self, value):
        self._product = value


class Extraction:
    @staticmethod
    def from_csv(file_path):
        sales = []
        with open(file_path, 'r', encoding='utf-8') as file:
            reader = csv.reader(file)
            next(reader)  # пропускаем заголовок
            for row in reader:
                sale_id = row[0]           # оставляем как строку
                date = row[1]
                amount = float(row[2])
                product = row[3]
                sale = Sale(sale_id, date, amount, product)
                sales.append(sale)
        return sales
```

#### **Задача 2**

Напишите класс `Transformation`, который содержит два метода: `filter_by_date(sales_data, start_date, end_date)` и `filter_by_amount(sales_data, min_amount, max_amount)`.

**Решение**

Напишите свое решение ниже

```python
import datetime

class Transformation:
    @staticmethod
    def filter_by_date(sales_data, start_date, end_date):
        # Преобразуем строковые даты в объекты date для сравнения
        start = datetime.datetime.strptime(start_date, '%Y-%m-%d').date()
        end = datetime.datetime.strptime(end_date, '%Y-%m-%d').date()
        filtered = []
        for sale in sales_data:
            sale_date = datetime.datetime.strptime(sale.get_date(), '%Y-%m-%d').date()
            if start <= sale_date <= end:
                filtered.append(sale)
        return filtered

    @staticmethod
    def filter_by_amount(sales_data, min_amount, max_amount):
        filtered = []
        for sale in sales_data:
            amount = sale.get_amount()
            if min_amount <= amount <= max_amount:
                filtered.append(sale)
        return filtered
```

#### **Задача 3**

Напишите класс `Analysis`, содержащий два метода: `calculate_total_sales` и `calculate_average_sales`. Возвращаемые значения округлите до двух знаков после запятой.

**Решение**

Напишите свое решение ниже

```python
class Analysis:
    @staticmethod
    def calculate_total_sales(sales_data):
        total = sum(sale.get_amount() for sale in sales_data)
        return round(total, 2)

    @staticmethod
    def calculate_average_sales(sales_data):
        if not sales_data:
            return 0.0
        total = sum(sale.get_amount() for sale in sales_data)
        average = total / len(sales_data)
        return round(average, 2)
```

#### **Задача 4**

Напишите класс `Loading` с методом `to_csv(sales_data, file_path)`, который запишет данные в `csv` файл.

**Решение**

Напишите свое решение ниже

```python
import csv

class Loading:
    @staticmethod
    def to_csv(sales_data, file_path):
        with open(file_path, 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(['id', 'date', 'amount', 'product'])
            for sale in sales_data:
                writer.writerow([
                    sale.get_id(),
                    sale.get_date(),
                    sale.get_amount(),
                    sale.get_product()
                ])
```
