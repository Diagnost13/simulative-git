#### **Задача 1**

Ваша задача написать функцию `get_time_to_conversion`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает среднее время между заходом на сайт и целевым действием.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res`

**Решение**

Напишите свое решение ниже

```python
import csv
from datetime import datetime
import ast

def parse_session(session_str):
    """Преобразует строку с сессией в кортеж (start_time, actions)."""
    data = ast.literal_eval(session_str)
    start_time_str = data[0]
    actions_raw = data[2]
    # Преобразуем время начала
    start_time = datetime.strptime(start_time_str, "%Y-%m-%d %H:%M:%S")
    # Преобразуем действия: каждая запись [page, time_str]
    actions = []
    for page, time_str in actions_raw:
        action_time = datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S")
        actions.append((page, action_time))
    return start_time, actions

def get_time_to_conversion(sessions):
    """
    Вычисляет среднее время (в секундах) от начала сессии до первого целевого действия "conversion".
    sessions: список сессий, каждая сессия представлена кортежем (start_time, actions),
              где start_time - datetime, actions - список кортежей (page, time).
    Возвращает среднее время, округлённое до двух знаков после запятой.
    """
    total_seconds = 0.0
    count = 0
    for start_time, actions in sessions:
        # Ищем первое действие с именем "conversion"
        for page, act_time in actions:
            if page == "conversion":
                delta = (act_time - start_time).total_seconds()
                total_seconds += delta
                count += 1
                break  # учитываем только первое conversion в сессии
    if count == 0:
        return 0.0
    avg = total_seconds / count
    return round(avg, 2)

# Чтение файла и подготовка данных
sessions_data = []
with open('sessions.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)  # пропускаем заголовок
    for row in reader:
        user_id, session_str = row
        try:
            start, actions = parse_session(session_str)
            sessions_data.append((start, actions))
        except Exception as e:
            # Если не удалось распарсить, можно пропустить или вывести предупреждение
            print(f"Ошибка парсинга строки: {row}, {e}")

# Вычисляем результат
res = get_time_to_conversion(sessions_data)

# Вывод для проверки (необязательно)
print(res)
```

#### **Задача 2**

Ваша задача написать функцию `get_session_duration`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает сколько вообще в среднем длится сессия.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res2`

**Решение**

Напишите свое решение ниже

```python
import csv
from datetime import datetime
import ast

def parse_session(session_str):
    """Преобразует строку с сессией в кортеж (start_time, duration, actions)."""
    data = ast.literal_eval(session_str)
    start_time_str = data[0]
    duration = data[1]                     # длительность сессии в секундах
    actions_raw = data[2]
    start_time = datetime.strptime(start_time_str, "%Y-%m-%d %H:%M:%S")
    actions = [(page, datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S")) for page, time_str in actions_raw]
    return start_time, duration, actions

# Загрузка данных
sessions_data = []
with open('sessions.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)  # пропускаем заголовок
    for row in reader:
        user_id, session_str = row
        try:
            start, duration, actions = parse_session(session_str)
            sessions_data.append((start, duration, actions))
        except Exception as e:
            print(f"Ошибка парсинга строки: {row}, {e}")

def get_session_duration(sessions):
    """
    Вычисляет среднюю длительность сессии (в секундах) на основе второго поля в записи сессии.
    sessions: список кортежей (start_time, duration, actions)
    Возвращает среднюю длительность, округлённую до двух знаков после запятой.
    """
    total_duration = sum(duration for _, duration, _ in sessions)
    count = len(sessions)
    if count == 0:
        return 0.0
    avg_duration = total_duration / count
    return round(avg_duration, 2)

res2 = get_session_duration(sessions_data)
print(res2)
```

#### **Задача 3**

Напишите функцию `get_daily_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает cколько разных заходов в течение дня делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res3`

**Решение**

Напишите свое решение ниже

```python
import csv
from datetime import datetime
import ast

def parse_session_with_user(row):
    """Парсит строку CSV и возвращает (user_id, start_time, duration, actions)."""
    user_id, session_str = row
    data = ast.literal_eval(session_str)
    start_time_str = data[0]
    duration = data[1]
    actions_raw = data[2]
    start_time = datetime.strptime(start_time_str, "%Y-%m-%d %H:%M:%S")
    actions = [(page, datetime.strptime(time_str, "%Y-%m-%d %H:%M:%S")) for page, time_str in actions_raw]
    return user_id, start_time, duration, actions

# Загрузка данных с user_id
sessions_with_user = []
with open('sessions.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)  # заголовок
    for row in reader:
        try:
            sessions_with_user.append(parse_session_with_user(row))
        except Exception as e:
            print(f"Ошибка парсинга строки: {row}, {e}")

def get_daily_session_count(sessions):
    """
    Вычисляет среднее количество сессий на один активный день пользователя.
    sessions: список кортежей (user_id, start_time, duration, actions)
    Возвращает среднее значение, округлённое до двух знаков после запятой.
    """
    total_sessions = 0
    unique_user_days = set()
    for user, start, _, _ in sessions:
        total_sessions += 1
        unique_user_days.add((user, start.date()))
    if not unique_user_days:
        return 0.0
    avg = total_sessions / len(unique_user_days)
    return round(avg, 2)

res3 = get_daily_session_count(sessions_with_user)
print(res3)
```

#### **Задача 4**

Напишите функцию `get_average_session_count`, которая принимает на вход сессии в формате, который вы определили для себя и подсчитывает сколько заходов вообще в среднем делает пользователь.

Результат - число типа `float`, округленное до двух знаков после запятой, сохраните в переменную `res4`

**Решение**

Напишите свое решение ниже

```python
def get_average_session_count(sessions):
    """
    Вычисляет среднее количество сессий на одного пользователя.
    sessions: список кортежей (user_id, start_time, duration, actions)
    Возвращает среднее значение, округлённое до двух знаков после запятой.
    """
    total_sessions = len(sessions)
    unique_users = len({user for user, _, _, _ in sessions})
    if unique_users == 0:
        return 0.0
    avg = total_sessions / unique_users
    return round(avg, 2)

res4 = get_average_session_count(sessions_with_user)
print(res4)
```
