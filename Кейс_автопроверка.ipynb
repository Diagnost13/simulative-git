```python
import importlib.util
import types

def get_first_function(module):
    """Возвращает первую определённую пользователем функцию в модуле."""
    for name in dir(module):
        if name.startswith('_'):
            continue
        attr = getattr(module, name)
        if isinstance(attr, types.FunctionType):
            return attr
    return None

def run_tests(standard_code_path, user_code_path, test_cases, output_file):
    """
    Сравнивает результаты эталонной программы и программы пользователя.

    Параметры:
        standard_code_path (str): путь к файлу с эталонным кодом
        user_code_path (str): путь к файлу с кодом пользователя
        test_cases (list): список входных данных. Если функция принимает несколько
                           аргументов, каждый элемент должен быть кортежем.
        output_file (str): имя файла для записи отчёта
    """
    # Загрузка эталонного модуля
    try:
        spec_std = importlib.util.spec_from_file_location("standard_module", standard_code_path)
        std_module = importlib.util.module_from_spec(spec_std)
        spec_std.loader.exec_module(std_module)
    except Exception as e:
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(f"Ошибка загрузки эталонного кода: {e}")
        return

    # Загрузка пользовательского модуля
    try:
        spec_user = importlib.util.spec_from_file_location("user_module", user_code_path)
        user_module = importlib.util.module_from_spec(spec_user)
        spec_user.loader.exec_module(user_module)
    except Exception as e:
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(f"Ошибка загрузки кода пользователя: {e}")
        return

    # Извлечение функций
    std_func = get_first_function(std_module)
    user_func = get_first_function(user_module)

    if std_func is None:
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write("Ошибка: в эталонном коде не найдено ни одной функции")
        return
    if user_func is None:
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write("Ошибка: в коде пользователя не найдено ни одной функции")
        return

    with open(output_file, 'w', encoding='utf-8') as f:
        for test_input in test_cases:
            # Вызов эталонной функции
            try:
                if isinstance(test_input, tuple):
                    expected = std_func(*test_input)
                else:
                    expected = std_func(test_input)
            except Exception as e:
                expected = f"Ошибка эталона: {e}"

            # Вызов пользовательской функции
            try:
                if isinstance(test_input, tuple):
                    actual = user_func(*test_input)
                else:
                    actual = user_func(test_input)
            except Exception as e:
                actual = f"Ошибка пользователя: {e}"

            # Запись только непройденных тестов
            if expected != actual:
                f.write("Тест не пройден\n")
                f.write(f"Ожидаемый результат: {expected}\n")
                f.write(f"Результат пользователя: {actual}\n")

# Пример использования в Google Colab
if __name__ == "__main__":
    # Пути к файлам (предполагается, что они загружены в среду)
    etalon_file = 'etalon.py'
    user_file = 'user.py'
    output_file = 'output.txt'

    # Тестовые случаи: для (1,2) → 3, (3,4) → 7, (5,6) → 11
    test_cases = [(1, 2), (3, 4), (5, 6)]

    # Запуск проверки
    run_tests(etalon_file, user_file, test_cases, output_file)
    print(f"Проверка завершена. Результаты сохранены в {output_file}")
```

#### **Задача 2**

Перед написанием кода, выполните следующую ячейку, в ней скачивается новый код пользователя, который проверит работу функции `check_for_malicious_code`.

**Решение**

Напишите свое решение ниже

```python
import ast
import importlib
import inspect

FORBIDDEN_NAMES = ['eval', 'exec']

def check_for_malicious_code(code, output_file):
    try:
        tree = ast.parse(code)
        for node in ast.walk(tree):
            if isinstance(node, ast.Attribute) and node.attr in FORBIDDEN_NAMES:
                raise ValueError("Код содержит запрещенный атрибут: {}".format(node.attr))
            elif isinstance(node, ast.Call) and isinstance(node.func, ast.Name) and node.func.id == 'eval':
                raise ValueError("Код содержит вызов eval()")
    except SyntaxError as e:
        with open(output_file, 'w') as f:
            f.write("Код содержит синтаксическую ошибку: {}\n".format(e.msg))
        return str(e)
    except ValueError as e:
        with open(output_file, 'w') as f:
            f.write("Код содержит вредоносные элементы: {}\n".format(str(e)))
        return str(e)

    with open(output_file, 'w') as f:
        f.write("Код не содержит вредоносные элементы.\n")

def run_tests(standard_code, user_code, test_cases, output_file):
    with open(output_file, 'w') as f:
        for test in test_cases:
            if isinstance(test, (int, float)):
                test = (test,)
            try:
                standard_result = standard_code(*test)
            except Exception as e:
                f.write("Ошибка при выполнении эталонного кода: {}\n".format(e))
                return

            try:
                user_result = user_code(*test)
            except Exception as e:
                f.write("Ошибка при выполнении кода пользователя: {}\n".format(e))
                return

            if standard_result == user_result:
                f.write("Тест пройден\n")
            else:
                f.write("Тест не пройден\n")
                f.write("Ожидаемый результат: {}\n".format(standard_result))
                f.write("Результат пользователя: {}\n".format(user_result))

# Чтение кодов из файлов
standard_module = importlib.import_module('etalon')
user_module = importlib.import_module('user_malicious')

test_cases = [(1, 2), (3, 4), (5, 6)]
output_file = 'output2.txt'
# Проверка на вредоносный код
malicious_code_message = check_for_malicious_code(inspect.getsource(user_module), output_file)
if malicious_code_message:
    with open(output_file, 'w') as f:
        f.write("Ошибка при проверке модуля user_malicious:\n")
        f.write(malicious_code_message)
else:
    run_tests(standard_module.add, user_module.add_numbers, test_cases, output_file)
```

Перед написанием кода, выполните следующую ячейку, в ней скачивается новый код пользователя, который проверит работу проверки на время выполнения.

```python
import ast
import importlib
import inspect
import time

FORBIDDEN_NAMES = ['eval', 'exec']

def check_for_malicious_code(code, output_file):
    try:
        tree = ast.parse(code)
        for node in ast.walk(tree):
            if isinstance(node, ast.Attribute) and node.attr in FORBIDDEN_NAMES:
                raise ValueError("Код содержит запрещенный атрибут: {}".format(node.attr))
            elif isinstance(node, ast.Call) and isinstance(node.func, ast.Name) and node.func.id == 'eval':
                raise ValueError("Код содержит вызов eval()")
    except SyntaxError as e:
        with open(output_file, 'w') as f:
            f.write("Код содержит синтаксическую ошибку: {}\n".format(e.msg))
        return str(e)
    except ValueError as e:
        with open(output_file, 'w') as f:
            f.write("Код содержит вредоносные элементы: {}\n".format(str(e)))
        return str(e)

    with open(output_file, 'w') as f:
        f.write("Код не содержит вредоносные элементы.\n")

import multiprocessing

def run_tests(standard_code, user_code, test_cases, output_file):
    with open(output_file, 'w') as f:
        for test in test_cases:
            if isinstance(test, (int, float)):
                test = (test,)
            try:
                standard_result = standard_code(*test)
            except Exception as e:
                f.write("Ошибка при выполнении эталонного кода: {}\n".format(e))
                return

            try:
                process = multiprocessing.Process(target=user_code, args=test)
                start_time = time.time()
                process.start()
                process.join(timeout=5)
                if process.is_alive():
                    process.terminate()
                    raise TimeoutError("Превышено время выполнения.")
                execution_time = time.time() - start_time
            except TimeoutError as e:
                f.write("Предупреждение: время выполнения превышает 5 секунд.")
                return
            except Exception as e:
                f.write("Ошибка при выполнении кода пользователя: {}\n".format(e))
                return

            if standard_result == user_result:
                f.write("Тест пройден\n")
            else:
                f.write("Тест не пройден\n")
                f.write("Ожидаемый результат: {}\n".format(standard_result))
                f.write("Результат пользователя: {}\n".format(user_result))


# Чтение кодов из файлов
standard_module = importlib.import_module('etalon')
user_module = importlib.import_module('user_inf')

test_cases = [(1, 2), (3, 4), (5, 6)]
output_file = 'output3.txt'
# Проверка на вредоносный код
malicious_code_message = check_for_malicious_code(inspect.getsource(user_module), output_file)
if malicious_code_message:
    with open(output_file, 'w') as f:
        f.write("Ошибка при проверке модуля user_malicious:\n")
        f.write(malicious_code_message)
else:
    run_tests(standard_module.add, user_module.add_numbers, test_cases, output_file)

```
